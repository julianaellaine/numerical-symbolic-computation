Question 3: Determining Survival Rate
- The script calculates the survival rate using df["Survived"].mean() and prints the result.
- It allows for an analysis of what the survival rate indicates.

```python
# Analyze Survival by Age
plt.figure(figsize=(10, 6))
plt.hist(df[df["Survived"] == 1]["Age"].dropna(), bins=20, alpha=0.5, label="Survived", color='blue')
plt.hist(df[df["Survived"] == 0]["Age"].dropna(), bins=20, alpha=0.5, label="Did Not Survive", color='red')
plt.legend()
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution of Survivors vs. Non-Survivors")
plt.show()
```

```python
# Compute average age of survivors and non-survivors
average_age_survivors = df[df["Survived"] == 1]["Age"].mean()
average_age_non_survivors = df[df["Survived"] == 0]["Age"].mean()

print("\nAverage Age of Survivors:", average_age_survivors)
print("Average Age of Non-Survivors:", average_age_non_survivors)

# Analysis of age impact on survival
if average_age_survivors < average_age_non_survivors:
    print("\nYounger passengers had a higher survival rate.")
else:
    print("\nOlder passengers had a higher survival rate.")
```
